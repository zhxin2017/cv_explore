```python
import sys
sys.path.append('..')
import torch
from torch import nn, Tensor
from torch.utils.data import Dataset, DataLoader
import torchvision
import random
import os
import cv2
from matplotlib import pyplot as plt
from od import detr_dataset as ds
from common.config import val_annotation_file, val_img_od_dict_file, val_img_dir, train_annotation_file, train_img_od_dict_file, train_img_dir, img_size
from od.config import cid_to_name
from torchvision import transforms
from od.box import cxcywh_to_xyxy
from od import detr_model, examine, anno, anchor, detr_model_base
import random
%matplotlib inline
```

```python
dicts_val = anno.build_img_dict(val_annotation_file, val_img_od_dict_file, task='od')
dicts_train = anno.build_img_dict(train_annotation_file, train_img_od_dict_file, task='od')
```

```python
device = torch.device("mps")
anchors = torch.tensor(anchor.generate_anchors(variable_stride=False), device=device)
n_query = 100
model = detr_model_base.DETR(d_cont=384, d_pos=128, d_head=64, n_enc_layer=16, n_dec_layer=8, n_query=n_query)
# model = detr_model2.DETR(d=448, d_img_coord_emb=64, d_anchor_coord_emb=32, n_head=7, d_indic=64, n_attn_layer=20,
#                          anchors=anchors, exam_diff=True)
model.to(device)
n_head = 8

model.load_state_dict(torch.load(f'/Users/zx/Documents/ml/restart/resources/od_detr_base_1.pt'))
extractor = examine.FeatureExtractor(model, layers=[m[0] for m in model.named_modules()])
```

```python
# for random val data, first decoder layer

data = ds.OdDataset(dicts_val, n_query=n_query, train=False, sample_num=10, random_shift=False)
dl = DataLoader(data, batch_size=1, shuffle=True)

for img, boxes_gt_xyxy, cids, _, img_id in dl:
    img, anchors_, boxes, names, attns, n_head, boxes_pred_xyxy, cls_pred = examine.examine_attn(img, extractor, n_head, device, 'dec_ca_layers.0.self_attn.q_proj', 'dec_ca_layers.0.self_attn.k_proj')
    examine.draw_attn(img, anchors_, boxes, names, attns, n_head)
    break
```

```python
# for random val data, last decoder layer

data = ds.OdDataset(dicts_val, n_query=n_query, train=False, sample_num=10, random_shift=False)
dl = DataLoader(data, batch_size=1, shuffle=True)

for img, boxes_gt_xyxy, cids, _, img_id in dl:
    img, anchors_, boxes, names, attns, n_head, boxes_pred_xyxy, cls_pred = examine.examine_attn(img, extractor, n_head, device, 'dec_ca_layers.7.self_attn.q_proj', 'dec_ca_layers.7.self_attn.k_proj')
    examine.draw_attn(img, anchors_, boxes, names, attns, n_head)
    break
```
